```python
import folium
from folium.plugins import MarkerCluster

# Initialize map
nasa_coordinate = [spacex_df['Lat'].mean(), spacex_df['Long'].mean()]
site_map = folium.Map(location=nasa_coordinate, zoom_start=6)

# Create a MarkerCluster object
marker_cluster = MarkerCluster().add_to(site_map)

# Loop through each row in the DataFrame
for index, record in spacex_df.iterrows():
    coord = [record['Lat'], record['Long']]
    
    # Set marker color based on success
    marker_color = 'green' if record['class'] == 1 else 'red'
    
    # Create and add marker
    marker = folium.Marker(
        location=coord,
        icon=folium.Icon(color='white', icon_color=marker_color, icon='rocket', prefix='fa'),
        popup=f"Launch Site: {record['Launch Site']}<br>Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    
    # Add to cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.plugins import MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# STEP 1: Enable mouse coordinates display
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

# Add to map
site_map.add_child(mouse_position)

# STEP 2: Define distance calculation function (Haversine formula)
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # Earth radius in km

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# STEP 3: Example coordinates for launch site and closest coast point
launch_coord = [28.5623, -80.5774]     # Example: CCAFS SLC-40
coast_coord = [28.563, -80.568]        # Example coast nearby

# Calculate distance
dist_km = calculate_distance(*launch_coord, *coast_coord)

# STEP 4: Add Marker to coast point and show distance as label
folium.Marker(
    location=coast_coord,
    icon=DivIcon(
        icon_size=(100, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(dist_km),
    )
).add_to(site_map)


```

```python
from folium.features import DivIcon

# Example proximity coordinates (replace with your actual data)
city_coord = [28.6100, -80.6000]
railway_coord = [28.5725, -80.5850]
highway_coord = [28.5635, -80.5700]
launch_coord = [28.5623, -80.5774]

# Calculate distances using the custom function
dist_city = calculate_distance(*launch_coord, *city_coord)
dist_railway = calculate_distance(*launch_coord, *railway_coord)
dist_highway = calculate_distance(*launch_coord, *highway_coord)

# Helper function to create distance marker and line
def add_distance_marker(site_map, start_coord, end_coord, distance_km, label_color='blue'):
    # Draw line between launch site and point
    folium.PolyLine(locations=[start_coord, end_coord], color=label_color, weight=2).add_to(site_map)
    
    # Add distance marker at the destination point
    folium.Marker(
        location=end_coord,
        icon=DivIcon(
            icon_size=(150, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:{};"><b>{:.2f} KM</b></div>'.format(label_color, distance_km),
        )
    ).add_to(site_map)

# Add markers and lines for each proximity point
add_distance_marker(site_map, launch_coord, city_coord, dist_city, 'purple')
add_distance_marker(site_map, launch_coord, railway_coord, dist_railway, 'orange')
add_distance_marker(site_map, launch_coord, highway_coord, dist_highway, 'green')

# Show the map
site_map
```
